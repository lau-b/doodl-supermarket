```python
import pandas as pd
```

```python
df = pd.read_csv('./data/raw/monday.csv', sep=';', index_col=0, parse_dates=True)
```

```python
df.head(40)
```

```python

resDF = df.groupby('customer_no').resample('1min').ffill()
resDF.head(20)
```

```python
resDF['next'] = resDF.groupby(resDF['customer_no']).shift(-1)
resDF
```

```python
resDF = resDF.rename(columns={'location': 'prev'})
resDF
```

```python
resDF.loc[resDF['customer_no']==1445]
```

```python
resDF.loc[(resDF['prev']=='checkout') & (resDF['next'].isna())]
```

```python
resDF.loc[(resDF['prev']!='checkout') & (resDF['next'].isna())]
```
